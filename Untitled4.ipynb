```python
from google.colab import files
import re

# Step 1: Upload the file
print("Upload your text file:")
uploaded = files.upload()

# Get the uploaded file name
input_file = list(uploaded.keys())[0]
output_file = "Aiken_Formatted.txt"

# Step 2: Function to convert to Aiken format
def convert_to_aiken(input_file, output_file):
    with open(input_file, 'r') as file:
        content = file.read().strip()  # Read the entire file as a single string

    # Split the content into individual questions using the pattern "Q XX:"
    questions = re.split(r'(Q \d+:)', content)

    aiken_lines = []
    question = ""
    options = []
    answer = ""

    for i in range(1, len(questions), 2):  # Process pairs of question number and content
        question_number = questions[i].strip()  # e.g., "Q 01:"
        question_content = questions[i + 1].strip()  # Content after "Q XX:"

        # Extract the question text and options
        main_question_match = re.match(r'^(.*?)(A: .*?)$', question_content, re.DOTALL)
        if main_question_match:
            question = main_question_match.group(1).strip()  # Extract main question
            options_block = main_question_match.group(2).strip()  # Extract options

            # Split the options and find the correct answer
            options = re.findall(r'([A-D]: .*?)(?=[A-D]:|$)', options_block)
            formatted_options = []
            answer = ""

            for option in options:
                clean_option = option.replace("**", "").strip()  # Remove "**" for correct answer
                formatted_options.append(clean_option.replace(':', '.'))  # AIKEN format
                if "**" in option:
                    answer = clean_option.split('.')[0]  # Extract "A", "B", etc.

            # Add to Aiken lines
            if question and formatted_options and answer:
                aiken_lines.append(question)
                aiken_lines.extend(formatted_options)
                aiken_lines.append(f"ANSWER: {answer}")
                aiken_lines.append("")  # Blank line between questions

    # Write the Aiken formatted file
    with open(output_file, 'w') as file:
        file.write("\n".join(aiken_lines))

    print(f"Aiken formatted file saved as: {output_file}")

# Step 3: Run the conversion
convert_to_aiken(input_file, output_file)

# Step 4: Download the output file
print("Download your Aiken formatted file:")
files.download(output_file)
```
